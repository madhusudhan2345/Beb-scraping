```python
iris_html=requests.get('https://en.wikipedia.org/wiki/Iris_flower_data_set')
print(iris_html)
```

```python
soup=BeautifulSoup(iris_html,'lxml')
print(soup)
```

```python
headers=soup.find('table')
headers_first=headers.find_all('th')
attributes=[]
for i in headers_first:
    attributes.append(i.text.replace(' ',''))
print(attributes)
```
